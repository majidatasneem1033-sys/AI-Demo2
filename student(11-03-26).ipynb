```python
import numpy as np
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt

df=pd.read_csv("D:\pyth1\student_data_clean.csv")

df.head()
```

```python
pd.pivot_table(df,
                 values=['CGPA',"Scholarship"],
                              index="Department",
                              columns="Gender",
                              aggfunc="mean" )
```

```python
df.columns
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df,x="Department",y="Scholarship")
plt.show()
```
