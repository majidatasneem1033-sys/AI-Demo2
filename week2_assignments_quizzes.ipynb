```python
import math

# 1. Get numbers from user
def get_numbers():
    user_input = input("Enter numbers separated by spaces: ")
    return [float(x) for x in user_input.split()]

# 2. Calculate mean
def calculate_mean(data):
    return sum(data) / len(data)

# 3. Calculate sample variance
def calculate_variance(data, mean):
    n = len(data)
    return sum((x - mean) ** 2 for x in data) / (n - 1)

# 4. Calculate standard deviation
def calculate_std_dev(variance):
    return math.sqrt(variance)


# Example usage
data = get_numbers()
mean = calculate_mean(data)
variance = calculate_variance(data, mean)
std_dev = calculate_std_dev(variance)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# 1. Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 2. Method to display info
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")


# 3. Derived class Student
class Student(Person):
    # 4. Add attributes student_id and major
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    # 5. Override display_info()
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, ID: {self.student_id}, Major: {self.major}")


# 6. Derived class GraduateStudent
class GraduateStudent(Student):
    # 7. Add thesis_topic
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    # Override display_info()
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, ID: {self.student_id}, "
              f"Major: {self.major}, Thesis: {self.thesis_topic}")


# 8. Create objects and call display_info()

p = Person("Ali", 30)
s = Student("Sara", 20, "S123", "Computer Science")
g = GraduateStudent("Ahmed", 25, "G456", "AI", "Machine Learning")

p.display_info()
s.display_info()
g.display_info()
```

```python
class Rectangle:
    # 1. Constructor
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # 2. __str__ method
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

    # 3. Area method
    def area(self):
        return self.width * self.height

    # 4. Overload + operator
    def __add__(self, other):
        return Rectangle(self.width + other.width, self.height + other.height)

    # 5. Overload == operator
    def __eq__(self, other):
        return self.area() == other.area()


# 6. Create rectangles
rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 3)

# 7. Print rectangles and areas
print(rect1)
print("Area:", rect1.area())

print(rect2)
print("Area:", rect2.area())

# 8. Add rectangles
rect3 = rect1 + rect2
print(rect3)
print("Area:", rect3.area())

# 9. Compare rectangles
print("rect1 == rect2:", rect1 == rect2)
print("rect1 == rect3:", rect1 == rect3)
```
