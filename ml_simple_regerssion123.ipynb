```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("D:\python\practice\data\student_marks.csv")
df.head()
```

```python
plt.scatter(df['Study_Hours'],df['Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
x=df[['Study_Hours']]#feature
y=df["Marks"]#target


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Training Data Shape:", x_train.shape,y_train.shape)
print("Testing Data Shape:",x_test.shape,y_test.shape)
```

```python
print("Training Data Shape:", x_train.shape,y_train.shape)
print("Testing Data Shape:",x_test.shape,y_test.shape)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

print("slop(wight)",model.coef_[0])
print("intercept(bias)",model.intercept_)
```

```python
plt.scatter(x_test,y_test,color='blue',label="Actual Marks")
plt.plot(x_test,y_pred,color='red',label="Regession line")
plt.xlabel("Study hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print("R² Score on Test Set:", r2 )
```

```python
y_train_pred = model.predict(x_train)
from sklearn.metrics import r2_score

r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_pred)

print("Training R²:", r2_train)
print("Testing R²:", r2_test)
```
