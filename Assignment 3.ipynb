```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Alice','Bob','Charlie','Diana','Eve','Frank'],
    'Age': [25,30,np.nan,35,28,np.nan],
    'Score': [85,92,78,np.nan,88,95],
    'Grade': ['A','A','B','C','B','A']
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
print(df.head(3))     
print(df.shape)       
print(df.columns)     
print(df.dtypes)     
```

```python
print(df.isnull())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
```

```python
df = df.dropna(subset=['Score'])
print(df)
```
