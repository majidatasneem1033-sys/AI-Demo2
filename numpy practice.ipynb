```python
import numpy as np
#convert list into array
arr=np.array([1,2,3])
print(arr)


#find the dimension of array
print(arr.ndim)

#type of array
print(type(arr))
```

```python
#numpy ones function
arr=np.ones(5)
print(arr)
```

```python
#make martix and only contain one block
arr=np.array([[4,8,2],[6,9,1],[3,7,5]])
print(arr)

#make martix using function
arr=np.ones((1,2))
print(arr)
```

```python
#change the data type using ones function
arr2=np.ones((3,5),dtype=int)
print(arr2)


# int into the bool
arr2=np.ones((3,4),dtype=bool)
print(arr2)

#bool into string
arr2=np.ones((3,4),dtype=str)
print(arr2)
```

```python
#numpy zero function
#same write but ones convert into the zeroo!
arr3=np.zeros((3,5))
print(arr3)

#find type
arr3=np.zeros((1,5))
print(type(arr3))

#change the type
arr3=np.zeros((3,5),dtype=str)
print(arr3)

arr3=np.zeros((3,4),dtype=bool)
print(arr3)

arr3=np.zeros((5,2),dtype=int)
print(arr3)
```

```python
# create empty martix
emp=np.empty((3,4))
print(emp)
```

```python
#marke array sequence number using arg method
arr3=np.arange(1,13)
print(arr3)
#1st number is included and last number is excluded

#increament numbers
arr3=np.arange(0,13,2)
print(arr3)


#some equal numbers between two numbers
#smaj naii ayiii
arr3=np.linspace(1,10,20)
print(arr3)
```

```python
#SHAPING an Array
import numpy as np
arr=np.array([[3,4],[1,3],[5,4],[8,4]])
print(arr.size)
print(arr.ndim)
print(arr)
#reshape use when small dimension convert into big dimension
arr=arr.reshape(2,4)
print(arr)
```

```python
#covert array big dimension into small dimension
#work on single block of array
arr=np.array([[[1,2,3],[3,4,5],[5,6,7]]])

#using ravel function

s=arr.ravel()

print(s)
```

```python
#Do Math Functions using NUMPY

#To add two arrays use this function:
arr1=np.array([[1,8,5],[4,6,1]])
arr2=np.array([[1,2,5],[2,4,7]])
np.add(arr1,arr2)


#to sub numbers
#arr1=np.array([[1,8,5],[4,6,1]])
#arr2=np.array([[1,2,5],[2,4,7]])
#np.subtract(arr1,arr2)


#To multiply two arrays use this function:
#arr1=np.array([[1,8,5],[4,6,1]])
#arr2=np.array([[1,2,5],[2,4,7]])
#np.multiply(arr1,arr2)


#For DOT multiply of two arrays use this function:
#np.dot() me matrix multiplication hoti hai.condition
#(columns of first matrix) = (rows of second matrix)
arr1=np.array([[1,8,5],[4,6,1]])
arr2=np.array([[1,2,5],[2,4,7]])

result = np.dot(arr1, arr2.T)
print(result)


#divide two arrays use this function
#arr1=np.array([[1,8,5],[4,6,1]])
#arr2=np.array([[1,2,5],[2,4,7]])
#np.divide(arr1,arr2)
```

```python
#find maximum number n row
arr=np.array([1,3,2,5,4,9])
print(arr.max(axis=0))

#find maximum number in colomn i
arr=np.array([[3,6,5],[7,4,1],[6,9,3]])
print(arr.max(axis=1))

#find minimum number in row
arr=np.array([[1,4,6],[8,4,0],[1,6,2]])
print(arr.min(axis=0))

#find minimum number in column
# agr axis use karna hi to 2d banii pere gii
arr=np.array([[1,3,4,8,34,99,12,90]])
print(arr.min(axis=1))
```

```python
#find that index at which maximum number
arr=np.array([[1,4,6],[8,4,0],[1,6,2]])
print(arr.max(axis=0))
print(np.argmax(arr))

#find that index at which minimum number
arr=np.array([[1,4,6],[8,4,0],[1,6,2]])
print(arr.min(axis=1))
print(np.argmin(arr))

#find square root of an array
arr=np.array([1,4,6,5,7,2,9,1])#smaj naii ayiii
print(np.sqrt(arr))
```
