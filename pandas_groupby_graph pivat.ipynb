```python
# DASHBOARD - Multiple business metrics
import pandas as pd
import matplotlib.pyplot as plt
st_df=pd.read_csv("D:\python\practice\data\student_data_clean.csv")
fig, axes =plt.subplots(2,2,figsize=(15,10))
## 1. Top-left
pivot2=st_df.groupby('Department')["CGPA"].sum().sort_values(ascending=False)
pivot2.plot(kind='bar',ax=axes[0,0],color='skyblue',edgecolor='black')
axes[0,0].set_title("Total CGPA by Department")
axes[0,0].set_xlabel('Department')
axes[0,0].set_ylabel("Total CGPA")

Att=st_df.groupby('Gender')['Attendance_Rate'].mean().sort_values()
Att.plot(kind='bar',ax=axes[0,1],color='salmon',edgecolor='black')
axes[0,1].set_title("Average attendance rate")
axes[0,1].set_xlabel("Gender")
axes[0,1].set_ylabel("Attendence Rate")

ss=st_df['Scholarship'].value_counts()
ss.plot(kind='pie',ax=axes[1,0],autopct='1%.1f%%')
axes[1,0].set_xlabel('Scholarship')
axes[1,0].set_title("Scholarship distribution")
axes[1,0].set_ylabel('Name')


sc=st_df.groupby('City')['Scholarship'].sum().sort_values()
sc.plot(kind='line',ax=axes[1,1],color='green',marker='o')
axes[1,1].set_title("Scholarship distribution by City")
axes[1,1].set_xlabel("City")
axes[1,1].set_ylabel("Total Scholarship")
```
