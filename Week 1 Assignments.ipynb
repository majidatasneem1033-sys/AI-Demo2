Assignment 3: The Number Cruncher (Day 5 Loops and Conditionals) 

```python
 #Use a for loop to iterate through numbers from 1 to 20.
for i in range(1, 21):
    print(i)

#2. For each number: 
   #- If divisible by 3, print "Fizz". 
   #- If divisible by 5, print "C". 
   #- If divisible by both 3 and 5, print "FizzBuzz". 
   #- Otherwise, print the number itself.'''
for i in range(1,21):
    if i%3==0:
       print("Fizz")
    elif i%5==0:
        print("buzz")
    else:
        print(i)

#3. After the loop, ask the user for a number.
user_number=int(input("Enter a number:"))
print(user_number)

#4. Use a while loop to print a countdown from that number down to 1, and then print "Blast Off!".
while user_number>0:
    print(user_number)
    user_number-=1
print("blast off!")

#5. Use break to stop the countdown if the user's number is greater than 10.
number=int(input("Enter a number:"))
while number>0:
    if number>10:
        print("number is greater than 10")
        break
    print(number)
```
