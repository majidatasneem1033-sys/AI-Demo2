```python
#SUM OF TWO NUMBERS
sum = lambda a,b : a+b
print(sum(2,3))
```

```python
#if number is even
sq = lambda n: n%2==0
print(sq(7))
```

```python
#mapping a function(mapping is basically use as iteration, it traverse for each value in functon )
def myfunc(a, b):
  return a + b

x = list(map(myfunc, ('apple', 'banana', 'cherry'), ('orange', 'lemon', 'pineapple')))
print(x)
```

```python
#returns if number is even if we actuallly want to print values not t/f
num=[1,2,3,5,6]
x=list(map(lambda n: n if n%2 !=0 else None,num))
print(x)
```
